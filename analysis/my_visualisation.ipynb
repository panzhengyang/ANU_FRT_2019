```python
import numpy as np
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt 
```

```python
mit = np.genfromtxt('cmit.txt')
ngl = np.genfromtxt('cngl.txt')
```

```python
flag1 = np.logical_and(mit[:,11]==0,ngl[:,11]==0)
flag2 = np.logical_and(mit[:,3]>180,ngl[:,3]>180)
```

```python
flag3 = abs(mit[:,3]-ngl[:,3]) < 20
```

```python
flag = flag1*flag2*flag3
```

```python
plt.axis([-15,15,-15,15])
```

```python
plt.plot([-300,300],[-300,300],'r')
```

```python
plt.scatter(mit[flag,7],ngl[flag,7])
```
